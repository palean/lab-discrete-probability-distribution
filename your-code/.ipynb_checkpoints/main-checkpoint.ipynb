## Bernoulli's Distribution

In a basket containing 100 fruits, there are 60 apples and 40 oranges. 

A fruit is picked randomly from the basket. 

1. **What is the probability that the fruit is apple?**
1. **What is the probability that it is an orange?**

````python
p = 60/100 # p: apples
q= 1-p

print(f'The probability that the fruit is an apple is {p}')
print(f'The probability that the fruit is an orange is {q}')
````

1. **What is the probability that the first 5 fruits are all apples and the next 15 fruits are all oranges?**

## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
#binom.pmf(k,n,p)
from scipy.stats import binom
````

````python
# your code here
p = .6 #apples
q = .4 #oranges
print(f'The probability that this time the sample contains 5 apples and 15 oranges is {round((binom.pmf(5,20,p) + binom.pmf(15,20,q))*100,3)}%')
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
# your code here

print(f'The probability that less than 5 fruits picked are apples is {round((binom.cdf(5,20,p))*100,3)}%')
````

**Plot the PDF (probability distributiom function) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
# your code here
# Please label the axes and give a title to the plot 
import matplotlib.pyplot as plt
import numpy as np

n = 20 #size
p = .6 #for apple

plt.figure(figsize=(22,12))
plt.plot(np.arange(0,20), binom.pmf(np.arange(0,20),n,p), 'ko', ms=10,alpha=.7);
plt.vlines(np.arange(0,20), 0, binom.pmf(np.arange(0,20),n,p), alpha=.5);
plt.title('Binomial probability distribution function', size=15)
plt.xlim(0,20);
plt.xlabel('Success', size=15);
plt.ylabel('Probability', size=15);
plt.xticks(np.arange(0,20));
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of 
```

````python
import math
math.factorial(n) 

# P (k goals in a match) = (2.3^k * e^-2.3) / k!
# P (k=5 goals in a match) = (2.3^5 * e^-2.3) / 5!
k=5
print(f'The probability that the number of goals scored in a match is 5: {round(((math.exp(-2.3)*(2.3**k))/(math.factorial(k)))*100,3)}%')
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
k = list(range(11))
p = [round(math.exp(-2.3)*(2.3**i/math.factorial(i)),3) for i in k]

plt.figure(figsize=(22,12))
plt.plot(k, p, 'ko', ms=10, alpha=0.7);
plt.vlines(k, 0, p, alpha=0.5);
plt.title('Poisson Probability Distribution Function',size=15);
plt.xlim(-1,11);
plt.xlabel('k goals in a match',size=15);
plt.ylabel('P(k) := probability of k goals in a match',size=15);
plt.xticks(np.arange(0,11));
````
